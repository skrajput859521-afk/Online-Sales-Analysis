```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:/Users/skraj/OneDrive/Desktop/New folder/Online Sales/Sample - Superstore.csv',encoding='latin-1')
df
df.shape[0]
```

```python
#Converting DATE COLUMNS
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
```

```python
df.head()
```

```python
#new column creation
df['Order Month']=df['Order Date'].dt.month
df['Order year']=df['Order Date'].dt.year
df['Order Day of week']=df['Order Date'].dt.dayofweek
df.head()
```

```python
#Monthly Sales Analysis

sales_by_month=df.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month=round(sales_by_month,2)
sales_by_month
```

```python

plt.figure(figsize=(12,6))

plt.plot(sales_by_month['Order Month'].astype(str), sales_by_month['Sales'], marker='o')

plt.title("Monthly Sales Trend", fontsize=16, fontweight='bold')
plt.xlabel("Order Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45)

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#category by sales

sales_by_Category=df.groupby('Category')['Sales'].sum().reset_index()
sales_by_Category=round(sales_by_Category,2)
sales_by_Category
```

```python

plt.figure(figsize=(8,6))

plt.bar(sales_by_Category['Category'], sales_by_Category['Sales'], color='blue', edgecolor='black')

plt.title("Sales by Category", fontsize=16, fontweight='bold')
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)

# Show values on top of bars
for i, v in enumerate(sales_by_Category['Sales']):
    plt.text(i, v + (v*0.01), str(v), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
sales_by_SubCategory=df.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_SubCategory=round(sales_by_SubCategory,2)
sales_by_SubCategory
```

```python

# Sort by Sales for better visualization
sales_by_SubCategory = sales_by_SubCategory.sort_values(by='Sales', ascending=False)

plt.figure(figsize=(12,6))

plt.bar(sales_by_SubCategory['Sub-Category'], sales_by_SubCategory['Sales'], 
        color='lightcoral', edgecolor='black')

plt.title("Sales by Sub-Category", fontsize=16, fontweight='bold')
plt.xlabel("Sub-Category", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45)

# Add values on top of bars
for i, v in enumerate(sales_by_SubCategory['Sales']):
    plt.text(i, v + (v*0.01), str(round(v,2)), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
Profit_by_month=df.groupby('Order Month')['Profit'].sum().reset_index()
Profit_by_month=round(Profit_by_month,2)
Profit_by_month
```

```python

# Data
sales_profit_by_segment = df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

# Plot
plt.figure(figsize=(8,5))
plt.bar(sales_profit_by_segment['Segment'], sales_profit_by_segment['Sales'], label='Sales')
plt.bar(sales_profit_by_segment['Segment'], sales_profit_by_segment['Profit'], label='Profit')

plt.title("Sales & Profit by Segment", fontweight='bold')
plt.xlabel("Segment", fontweight='bold')
plt.ylabel("Amount", fontweight='bold')
plt.legend()
plt.show()
```

```python
# Sales to Profit ratio by Segment
sales_profit_by_segment = df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

# Create new column
sales_profit_by_segment['sales_to_profit_ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']

# Print selected columns
print(sales_profit_by_segment[['Segment','sales_to_profit_ratio']])
```

```python
import matplotlib.pyplot as plt

# Sales & Profit Aggregation
sales_profit_by_segment = df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

# Sales to Profit Ratio
sales_profit_by_segment['sales_to_profit_ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']

# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(
    sales_profit_by_segment['sales_to_profit_ratio'],
    labels=sales_profit_by_segment['Segment'],
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Sales-to-Profit Ratio by Segment", fontweight='bold')
plt.show()
```
